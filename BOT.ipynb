**2. Pattern Recognition:**

- Tên dạng bài toán : Bài toán quy hoạch động
- Kĩ thuật áp dụng, đặc điểm nhận dạng: Mảng con lớn nhật tại một vị trí được tính toán đơn giản từ các bài toán con nhỏ hơn là mảng con lớn nhất tại vị trí trước đó  Thuật toán Kadane’s
Thuật toán Kadane’s quét qua các phần tử của mảng, tính tại từng vị trí tổng lớn nhất có thể của các mảng con kết thúc tại vị trí đó. Mảng con này có thể rỗng khi tổng của các phần tử bằng 0 hoặc có thể chứa nhiều hơn một phần tử so với mảng con lớn nhất tại vị trí trước đó.

**4. Đánh giá độ phức tạp**

Sử dụng phương pháp trên, chúng ta cần lặp lại mảng chỉ một lần, điều này tốt hơn rất nhiều so với các cách tiếp cận khác. Hay nói chính xác hơn, độ phức tạp về thời gian của Thuật toán Kadane là O(n)

**5.Programming.**



```
n = int(input())
A = list(map(int, input().split()))
a = s = 0
start = end = current = 0
for i in range(n):
  if a+A[i] < A[i]:
    current = i
    a = A[i]
  else:
    a += A[i]
  if a>s:
    s = a
    start = current
    end = i
print(start+1, end+1, s)

```
