#**Problem 2: Phân số**
#**1. Abstraction:**
Với các giá trị a,b,c,d cho trước, lần lượt tăng a,b lên 1 đơn vị, tính số lần tăng cần thiết để a=c và b=d
#**2. Pattern Recognition:**
Thuật toán sử dụng : Brute Force

Đặc điểm nhận dạng: Vì bài toán phải xét từng lượt tăng để kiểm tra cho đến khi thỏa điều kiện nên em xài brute force
#**3. Algorithm:**
Tăng lần lượt a,b lên và rút gọn. Kiểm tra nếu a/b = c/d thì xuất biến đếm (hoặc 0 nếu a/b != c/d) và kết thúc
 
Sử dụng hàm gcd trong thư viện math để rút gọn a,b sao khi tăng để tối ưu kết quả
image.png
#**4. Độ phức tạp và Code:**

Độ phức tạp trung bình : O(n)

```python
import math
a=int(input())
b=int(input())
c=int(input())
d=int(input())
dem=0
while (a/b < c/d):
  dem+=1
  a=a+1
  b=b+1
  x=math.gcd(int(a),int(b))
  a=a/x
  b=b/x
if (a/b == c/d):
  print(dem)
else:
  print(0)
```
